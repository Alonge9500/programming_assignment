##### Cleaning
* Removing the outliers using Interquatile range method

```python
# Calculate the 75th percentile of the Salary column
q3 = df['Salary'].quantile(0.75)
iqr = q3 - df['Salary'].quantile(0.25)
upper_bound = q3 + 1.5 * iqr
df = df[df['Salary'] <= upper_bound]
```

```python
def plot_topn_category(category,n):
    
    top_categories = categorical_df[category].value_counts().nlargest(n).index.tolist()

    # Create a new dataframe with only the top 10 categories
    df_top_categories = categorical_df[categorical_df[category].isin(top_categories)]

    # Plot the count plot
    sns.countplot(x=category, data=df_top_categories, order=top_categories)

    plt.title(f'Top {n} Categories with the Highest Count for {category}')
    plt.xlabel(category)
    plt.ylabel('Count')
    plt.xticks(rotation=90)

    # Show the plot
    plt.show()
    
```

```python
plot_topn_category('Designation',15)
```

#### Observation
* Of all the Designation in the Dataset the top 15 designation are as displayed above and the designation with the highest count is the software engineer followed by the software developers

```python
plot_topn_category('Specialization',15)
```

```python
sns.countplot(data=categorical_df, x='Degree')
plt.show()
```

#### Observation

```python
def plot_topn_category_numerical(category,numerical,n,type_plot = 'box'):
    
    top_categories = df[category].value_counts().nlargest(n).index.tolist()

    # Create a new dataframe with only the top n categories
    df_top_categories = df[df[category].isin(top_categories)]

    # Plot 
    if type_plot == 'box':
        sns.boxplot(x=category, y=numerical,data=df_top_categories)
    if type_plot == 'vio':
        sns.violinplot(x=category, y=numerical,data=df_top_categories)
        

    plt.title(f'{numerical} vs {category}')
    plt.xlabel(category)
    plt.ylabel(f'{numerical}')
    plt.xticks(rotation=90)

    # Show the plot
    plt.show()
    
```

```python
plot_topn_category_numerical('Designation','Salary',15)
```

#### Observation 
* There are bits of changes in the salary base on the Designation 
* On Average Senior Software Engineer seems to earn the highest among top 15 Designation while the software developer tends to earn lower

```python
plot_topn_category_numerical('Specialization','Salary',15)
```

#### Observation
* Those that specialize in Computer engineering tends to earn better than others among the top 15
